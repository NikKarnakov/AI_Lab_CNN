# Функция для отображения изображений

Напишем функцию, которая будет использоваться для отображения изображений, выбранных из датасета. Она создает матрицу изображений и выводит их на экран.

```python
import matplotlib.pyplot as plt
import os
from PIL import Image
import numpy as np

def display_images(l, titles=None,fontsize=50, bs = 5):
    n = min(bs, 5)
    fig,ax = plt.subplots(1,n)
    for i,im in enumerate(l):
        if i == n:
          break
        ax[i].imshow(im)
        ax[i].axis('off')
        if titles is not None:
            ax[i].set_title(titles[i],fontsize=fontsize)
    fig.set_size_inches(fig.get_size_inches() * n)
    plt.tight_layout()
    plt.show()
```

# Аугментация данных и генерация батчей

`ImageDataGenerator` — это класс, который позволяет загружать и изменять изображения в реальном времени с аугментацией данных. Аугментация включает изменения таких параметров, как повороты, горизонтальные перевороты, масштабирование и изменение яркости, чтобы улучшить обобщающую способность модели.

Здесь создаются два генератора:

* `data_gen_aug` — для тренировочных данных с аугментацией (случайные изменения изображений);
* `data_gen` — для тестовых данных (без аугментации).

Изображения загружаются из папки `/content/petfaces`, и используется 80% для обучения, 20% — для проверки.

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

batch_size = 5
img_size = (128, 128)

data_gen_aug = ImageDataGenerator(rescale=1./255.,
                                  validation_split=0.2,
                                  rotation_range=30,
                                  horizontal_flip=True,
                                  zoom_range=0.2,
                                  brightness_range=[0.5, 1])

data_gen = ImageDataGenerator(rescale=1./255., validation_split=0.2)

train_data = data_gen_aug.flow_from_directory("/content/petfaces",
                                              batch_size=batch_size,
                                              target_size=img_size,
                                              subset="training",
                                              seed=20,
                                              class_mode='categorical',
                                              shuffle=True)

test_data = data_gen.flow_from_directory("/content/petfaces",
                                         batch_size=batch_size,
                                         target_size=img_size,
                                         subset="validation",
                                         seed=20,
                                         class_mode='categorical',
                                         shuffle=False)
```

# Визуализация тренировочных данных

Функция `display_images` отображает несколько изображений из тренировочного набора вместе с метками классов.

```python
display_images(train_data[0][0], np.argmax(train_data[0][1], axis = 1), batch_size)
```

# Создание модели нейронной сети

Этот фрагмент создает модель с использованием последовательной архитектуры `Sequential`. Структура модели включает:

* `Сверточные слои (Conv2D)` — для извлечения признаков из изображений. Первый сверточный слой имеет 32 фильтра размером 15x15, второй — 64 фильтра размером 5x5, а третий — 128 фильтров размером 3x3;
* `MaxPooling2D` — слои подвыборки для уменьшения размерности признаков;
* `Dense` слои — полностью связанные слои. Используются функции активации `relu` и `tanh`;
* `Flatten` — преобразует многомерные данные в одномерные для подачи в Dense-слои;
* `Softmax`-слой на выходе — для многоклассовой классификации.


```python
from keras.layers import *

model = keras.models.Sequential()

model.add(Conv2D(32, (15, 15), input_shape=(128, 128, 3), activation='tanh'))
model.add(MaxPooling2D())
model.add(Conv2D(64, (5, 5), activation='tanh'))
model.add(MaxPooling2D())
model.add(Dense(128,  activation="relu"))
model.add(Dense(128,  activation="tanh"))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D())
model.add(Flatten())
model.add(Dense(1024,  activation="tanh"))
model.add(Dense(1024,  activation="tanh"))
model.add(Dense(512,  activation="tanh"))
model.add(Dense(35, activation='softmax'))

model.summary()
```

# Компиляция модели

Модель компилируется с использованием оптимизатора `Adam`, функции потерь `categorical_crossentropy` и метрик точности (`accuracy` и `TopKCategoricalAccuracy)`, где `TopK` измеряет, была ли правильная метка в топ-5 предсказанных классов. `model.summary()` выводит краткую информацию о слоях модели.

```python
from tensorflow.keras.metrics import TopKCategoricalAccuracy

# Компиляция с метриками
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy', TopKCategoricalAccuracy(k=5)])
```

# Тренировка модели

Модель тренируется на тренировочном наборе данных. Валидация проводится на тестовом наборе. Тренировка длится 15 эпох.

```python
hits = model.fit(train_data,
                 validation_data=test_data,
                 epochs=15)
```

# Визуализация результатов обучения

Отображение графиков изменения точности на тренировочных и тестовых данных по эпохам.

```python
plt.plot(hits.history['accuracy'])
plt.plot(hits.history['val_accuracy'])
```

График демонстрирует изменение точности на обучающей и валидационной выборках на протяжении эпох. Видно, что кривые нестабильны, что может свидетельствовать о проблемах с обучением модели, таких как переобучение или недостаточный объем данных. Важно отметить, что стабильный рост точности не наблюдается, особенно на валидационной выборке.

# Оценка модели

Модель предсказывает классы для тестового набора данных. Далее, используя метрику точности `Accuracy`, вычисляется общая точность модели на тестовых данных.

```python
n_batches = len(test_data)

test_labels = np.concatenate([np.argmax(test_data[i][1], axis = 1) for i in range(n_batches)])
predictions = np.argmax(model.predict(test_data, steps = n_batches), axis=1)


m = tf.keras.metrics.Accuracy()
m.update_state(test_labels, predictions)
m.result().numpy()
```

Это крайне низкий показатель точности, свидетельствующий о том, что модель практически не может отличить классы друг от друга.

# Преобразование для бинарной классификации

Тестовые метки и предсказания преобразуются в бинарную классификацию: классы 0-11 соответствуют кошкам, 12 и выше — собакам.

```python
binary_test_labels = test_labels.copy()
binary_test_labels[binary_test_labels < 12] = 0
binary_test_labels[binary_test_labels >= 12] = 1

binary_predictions = predictions.copy()
binary_predictions[binary_predictions < 12] = 0
binary_predictions[binary_predictions >= 12] = 1


m = tf.keras.metrics.Accuracy()
m.update_state(binary_test_labels, binary_predictions)
m.result().numpy()
```

 В бинарной классификации, где классы объединены (кошки против собак), точность значительно лучше — 65.4%. Это указывает на то, что модель лучше различает между этими двумя большими категориями, но не может корректно классифицировать более мелкие подкатегории.

# Матрица ошибок

Создается и визуализируется матрица ошибок для оценки того, какие классы модель путает чаще всего.

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay

n_batches = len(test_data)
lablesNames = ['cat_Abyssinian', 'cat_Bengal', 'cat_Birman', 'cat_Bombay',
               'cat_British', 'cat_Egyptian', 'cat_Maine', 'cat_Persian',
               'cat_Ragdoll', 'cat_Russian', 'cat_Siamese', 'cat_Sphynx',
               'dog_american', 'dog_basset', 'dog_beagle', 'dog_boxer',
               'dog_chihuahua', 'dog_english', 'dog_german', 'dog_great',
               'dog_havanese', 'dog_japanese', 'dog_keeshond', 'dog_leonberger',
               'dog_miniature', 'dog_newfoundland', 'dog_pomeranian', 'dog_pug',
               'dog_saint', 'dog_samoyed', 'dog_scottish', 'dog_shiba',
               'dog_staffordshire', 'dog_wheaten', 'dog_yorkshire']


test_labels = np.concatenate([np.argmax(test_data[i][1], axis = 1) for i in range(n_batches)])
predictions = np.argmax(model.predict(test_data, steps = n_batches), axis=1)

cm = confusion_matrix(test_labels, predictions)
disp = ConfusionMatrixDisplay(confusion_matrix = cm,
                               display_labels = lablesNames)

fig = plt.figure(figsize = (10, 10))
ax = fig.add_axes([0, 0, 1.5, 1.5])

disp.plot(ax = ax, xticks_rotation='vertical')
plt.show()
```

Матрица ошибок показывает, как модель справляется с классификацией отдельных классов (разные породы кошек и собак). На матрице можно заметить сильную концентрацию в одном столбце для классов, что указывает на то, что модель ошибочно предсказывает большинство изображений в одну категорию — это явно указывает на проблему перекоса данных или неправильное обучение модели. Например, собаки предсказаны как "dog_chihuahua" 36 раз, хотя другие классы не были предсказаны.

# Оценка топ-3 точности

Проводится оценка по метрике Top-3 точности: модель считается успешной, если правильный класс оказался среди трёх наиболее вероятных.

```python
r = 0

test_labels = np.concatenate([np.argmax(test_data[i][1], axis = 1) for i in range(n_batches)])
predictions = np.argmax(model.predict(test_data, steps = n_batches), axis=1)

for i in range(len(test_labels)):
  if (test_labels[i] <= 12 and predictions[i] <= 12) or (test_labels[i] > 12 and predictions[i] > 12):
      r += 1

print("acc = {}".format( r / len(test_labels)))

r = 0

predictions = model.predict(test_data, steps = n_batches)
indexes = np.argsort(-predictions, axis = 1)

for i in range(len(predictions)):
  if (((indexes[i][0] <= 12 or indexes[i][1] <= 12 or indexes[i][2] <= 12) and (test_labels[i] <=12)) or (indexes[i][0] > 12 or indexes[i][1] > 12 or indexes[i][2] > 12) and (test_labels[i] > 12)):
    r += 1;

print("top 3 acc = {}".format( r / len(test_labels)))
```

Здесь мы видим, что модель достигает 100% точности при учете трех наиболее вероятных предсказаний. Это подтверждает, что модель включает правильный класс в свои топ-3 предсказания, хотя первое предсказание может быть неверным.
